### 2. Inferring population mean from sample - Confidence Intervals

Suppose we want to determine the mean value of a population parameter and we have drawn a simple random sample of size $n$ from that population.

The simple random sample can be measured and its mean and standard deviation are known, $\bar{x}$ and $S$ respectively<br>
We might be tempted to assume that the population mean is the same as the mean of the sample that we have drawn. However we know that sampling introduces certain random errors known as sampling error. Hence it would make sense to quantify the confidence with which we claim to know the population mean. Also, given the random error prone nature of the process, we can not make precise point estimates of the population mean. This is why we talk in terms of intervals within which we believe the population parameter to lie.<br><br>
This is where confidence intervals come in. Confidence intervals make use of the central limit theorem (CLT) to make statements about the population parameter based on properties of a simple random sample frawn from that population.<br>

Since we have a simple random sample of size $n$, with sample mean $\bar{x}$ and sample standard deviation $\sigma_\bar{x}$. We know from the CLT that the sampling distribution of sample means of samples with size $n$ has mean and standard deviation given by:

$
\begin{align}
\mu_\bar{x} = \mu
\end{align}
$<br>

$
\begin{align}
\sigma_\bar{x} = \frac{\sigma}{\sqrt{n}}
\end{align}
$<br>

We know that the sample standard deviation $S$ is an unbiased estimator fo the population standard deviation $\sigma$ and has very little variance of its own for sufficiently large samples. Hence we will assume the unknown population standard deviation $\sigma$ to be equal to the sample standard deviation $S$. Hence we can restate the sample standard deviation as:<br>

$
\begin{align}
\sigma_\bar{x} = \frac{S}{\sqrt{n}}
\end{align}
$<br>




We also need to pre decide, what is the level of significance $\alpha$ at which we want to state the confidence interval. Here $\alpha$ represents the probability of making an error. The parameter $1-\alpha$ represents the confidence level of the confidence interval. It is typical to choose confidence levels such as 0.05 (lower confidence) or 0.01 (higher confidence)

from our sampling distribution we determine the z values corresponding to the significance level (using a statistical table or using R's `qnorm` function. The interval defined by these z values will capture the observed sample mean $\bar{x}$ with probability $1-\alpha$. i.e.<br>

$P[\mu-z_{\frac{\alpha}{2}} \le \bar{x} \le \mu+z_{\frac{\alpha}{2}}] = 1- \alpha$

</div>



<br> Since $\bar{x}$ is the unknown here. We can rewrite the interval by subtracting $\mu$ and $\bar{x}$ throughout<br>
$ \Rightarrow P[\mu-z_{\frac{\alpha}{2}}-(\bar{x} + \mu) \le \bar{x}-(\bar{x} + \mu) \le \mu+z_{\frac{\alpha}{2}}-(\bar{x} + \mu)] = 1- \alpha$<br>
$ \Rightarrow P[-\bar{x}-z_{\frac{\alpha}{2}} \le - \mu \le -\bar{x}+z_{\frac{\alpha}{2}}] = 1- \alpha$<br>

multiplying by -1 inside the inequality. Which will flip the inequality symbols<br>
$ \Rightarrow P(x+z_{\frac{\alpha}{2}} \ge \mu \ge x-z_{\frac{\alpha}{2}}) = 1- \alpha$<br>

Rewriting the inequality from left to right as per convention<br>
$ \Rightarrow P(x-z_{\frac{\alpha}{2}} \le \mu \le x+z_{\frac{\alpha}{2}}) = 1- \alpha$<br>

This inequality is our confidence interval at the significance level $\alpha$<br>
$x-z_{\frac{\alpha}{2}} \le \mu \le x+z_{\frac{\alpha}{2}}$<br>

It means that when we draw a sample of size n from a population. The true population mean will be between $x-z_{\frac{\alpha}{2}}$ and $x+z_{\frac{\alpha}{2}}$ with probability $1-\alpha$.

$CI_\alpha: x\pm z_{\frac{\alpha}{2}}$

Note that we can reframe the concept as a hypothesis test (more on hypothesis testing following). Where the null hypothesis $H_0$ and alternative hypothesis $H_1$ can be stated as:

$H_0: \mu = \bar{x}$<br>
$H_1: \mu \ne \bar{x}$<br>
